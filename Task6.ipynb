```python
def getCurrencyRates(price,currency):
    qurl = 'https://api.coinbase.com/v2/exchange-rates'
    param = {
        'currency' : currency
    }
    ask_currency = input("what currency you want? : ")
    currencies = rq.get(url=qurl,params=param)
    currency_rate = float(currencies.json()['data']['rates'][ask_currency])
    result = currency_rate * price
    return ask_currency , result

```

```python
stock_new_list = []
for stock in stockdata:
    print(stock)
    change , rate = getCurrencyRates(price=stock['price'] , currency= stock['currency'])
    stock['to_change'] = change
    stock['exchange_rate'] = rate
    print(stock)
    stock_new_list.append(stock)

print(stock_new_list)

```
